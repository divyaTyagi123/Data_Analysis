```python
import seaborn as sns
import pandas as pd
import cufflinks as cf
from plotly.offline import iplot
```

```python
tips = sns.load_dataset("tips")

# ------------------------------------------------------------
# STEP 4: Enable offline plotting
# ------------------------------------------------------------
cf.go_offline()
cf.set_config_file(theme='pearl', sharing='public', offline=True)
```

```python
tips['total_bill'].iplot(
    kind='line',
    colors=['rgba(55, 128, 191, 1.0)'],  # ✅ valid color string
    title='Total Bill Trend (Cufflinks + Plotly)',
    xTitle='Index',
    yTitle='Total Bill ($)',
)
```
