```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
# plot makes line graph having x axis and y axis
plt.plot(x,y)
plt.title("Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

Subplots in matplotlib

```python
# it takes rows, columns and select a subplot
plt.subplot(2,2,1)
plt.plot(x,x)

plt.subplot(2,2,2)
plt.plot(x**2, x**2)

plt.subplot(2,2,3)
plt.plot(y,x)

plt.subplot(2,2,4)
plt.plot(x,y)
```

Object Oriented way


Matplotlib allows the aspect rratio, DPI and figure size to be specified when the 
figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch(pixel-per-inch)

```python
# fig is nothing initially , we have to add whatever we want to show and how we want to shape our canvas
# left , bottom , width , height
# in this you have a object which can be used anywhere
# you don't have a strict subplot area you can plot a sublot wherever you want and what size you want
```

```python
fig = plt.figure(figsize = (20,8),dpi = 50)      # dpi enhance clearity
axis1 = fig .add_axes([0.1,0.1,0.8, 0.8]) 
axis1.plot(x,y)

axis2 = fig .add_axes([0.5,1,0.6, 0.6])
axis2.plot(x,y)
```

```python
xaxis = np.linspace(0,5,11)
```

```python
fig , ax = plt.subplots(figsize = (12,6))

ax.plot(xaxis, xaxis + 1 , color= "red" , linewidth = 0.25)
ax.plot(xaxis, xaxis + 2 , color= "red" , linewidth = 0.5)
ax.plot(xaxis, xaxis + 3 , color= "red" , linewidth = 1.00)
ax.plot(xaxis, xaxis + 4 , color= "red" , linewidth = 2)

# possible linestyle options '-' , '-.' ,'--' ':', 'steps'
ax.plot(xaxis, xaxis+5 , color = "green", lw = 3, linestyle = '-')
ax.plot(xaxis, xaxis+6 , color = "green", lw = 3, linestyle = '-.')
ax.plot(xaxis, xaxis+7 , color = "green", lw = 3, linestyle = ':')

# custom dash
(line,) = ax.plot(xaxis, xaxis+8 , color="black", lw = 1.50)     # unpack the line w
line.set_dashes([5,50,15,50])   # format: line length, space length, ....

# possiblr marker symbols: marker = '+', 'o', 's', ',' , '.' , '1' , '2', ' 3' ......
ax.plot(xaxis, xaxis+9 , color = "blue", lw= 3, ls = '-' , marker= '+')
ax.plot(xaxis, xaxis+10 , color = "blue", lw= 3, ls = '-' , marker= 'o')
ax.plot(xaxis, xaxis+11 , color = "blue", lw= 3, ls = '-' , marker= 's')
ax.plot(xaxis, xaxis+12 , color = "blue", lw= 3, ls = '--' , marker= '1')
ax.plot(xaxis, xaxis+13 , color = "blue", lw= 1, ls = '-' , marker= 'o', markersize = 2)
ax.plot(xaxis, xaxis+14 , color = "blue", lw= 1, ls = '--' , marker= 'o', markersize = 4)
ax.plot(xaxis, xaxis+15 , color = "blue", lw= 1, ls = '-' , marker= 'o', markersize = 8, markerfacecolor = "red")
ax.plot(xaxis, xaxis+16 , color = "blue", lw= 1, ls = '-' , marker= 's', markerfacecolor = "yellow", markeredgewidth = 3, markeredgecolor = "green")
```
