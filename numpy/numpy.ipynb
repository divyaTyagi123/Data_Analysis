```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
b = np.sum(s , axis = 1)
b == 45
for i in b:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("It is valid for rows")
```

```python
c = np.sum(s , axis = 0)
c == 45
```

```python
a = s[0:3,0:3]
a.sum() == 45
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        a = s[i: i+3 , j:j+3]
        print(a.sum() == 45)
        
```

```python
#Column : [Age,Maths marks , Science marks]
data = np.array([
    [18, 85, 78],
    [19, 92, 88],
    [17, 76, 95],
    [18, 65, 70],
    [20, 90, 85]
])
```

```python
# Get the shape of the matrix
data.shape
```

```python
# find the average age of students.
# d = data[0: , 0:1]
# d.mean()
np.mean(data[:, 0])
```

```python
# extract maths marks of all students 
print(data[:,1])
```

```python
# find the highest science marks
np.max(data[:, 2])
```

```python
# get details of the student who scored more than 90 in math
data[data[: ,1] > 90]
```

```python
data[: , 1] + 5
```

```python
# find how many students are younger than 19
np.sum(data[: , 0] < 19, axis=0)
```

```python
# calculate the average marks in each subject (column-wise mean)
np.mean(data[:, 1:] , axis = 0)
```

```python
# get data of students who scored atleast 80 in both subjects.
data[(data[: , 1] >= 80) & (data[: , 2] >= 80)]
```

```python
# replace all science marks < 75 with 0
data[: , 2][data[:, 2] <75] =0
data
```
