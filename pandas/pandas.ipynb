# Pandas Series -
A series is one dimesional labeled array capable of holding any data type. the axis labels are collectively called the index.

# Creating Series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a' , 'b' , 'c']
my_list = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {1:10 , 2:20 , 3:30}
```

```python
pd.Series(my_list)
```

```python
# 0, 1 , 2 are labels
# 10, 20, 30 are datapoints
```

```python
pd.Series(my_list, index = labels)
```

```python
pd.Series(arr , index = labels)
```

```python
pd.Series(d)
```
