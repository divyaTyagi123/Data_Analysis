Feature Extraction

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ')':
            break
        if check == True:
            data += i
        if i == '(':
            check = True
    return data   
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
df["Episodes"] = df['Episodes'].str.replace(" eps","")
df
```

```python
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data += txt[j]
            return data
```

```python
df['Total Time'] = df['Title'].apply(extraction_time)
```

```python
from datetime import datetime

# Function to calculate total months
def get_total_months(time_range: str) -> int:
    try:
        start_str, end_str = time_range.split(" - ")

        # Convert to datetime (assuming format "Mon YYYY")
        start_date = datetime.strptime(start_str, "%b %Y")
        end_date = datetime.strptime(end_str, "%b %Y")

        # Calculate difference in months
        total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
        return total_months
    except:
        return None
df['Duration_Months'] = df['Total Time'].apply(get_total_months)
df
```
