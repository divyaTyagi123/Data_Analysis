```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()         # deck has a lot of null values thats why we have to drop as no mean , median or mode will be useful to fill this nan data points 
```

```python
df.drop(['deck' , 'embark_town' , 'alive' , 'class', 'who' , 'adult_male'] , axis = 1 , inplace = True)
```

```python
df['age'].fillna(df['age'].mean() , inplace = True)
```

```python
df.dropna(subset = ['embarked'] , inplace = True)       # drop rows having null value
```

```python
df.info()
```

```python
# Label Encoding 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked']) 
```

```python
df = df.astype(int)
df.head()
```

```python
x = df.drop('survived' , axis = 1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

# logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test , y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_test))
```

# KNN Model 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors = 5)
knn_model.fit(x_train_scaled , y_train)
```

```python
y_pred_knn = knn_model.predict(x_test_scaled)
```

```python
accuracy_score(y_test, y_pred_knn)
```

```python
# logistic regresion give 80% score and knn gives 78% 
```

```python
confusion_matrix(y_test , y_pred_knn)
```

```python
print(classification_report(y_test , y_pred_knn))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_NB = GaussianNB()
model_NB.fit(X_train,y_train)
```

```python
y_pred_NB = model_NB.predict(X_test)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DT = DecisionTreeClassifier(random_state = 42)
```

```python
model_DT.fit(x_train_scaled, y_train)
```

```python
y_pred_DT = model_DT.predict(x_test_scaled)
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(kernel = 'rbf')
```

```python
model_svm.fit(x_train_scaled , y_train)
```

```python
y_pred_svm = model_svm.predict(x_test_scaled )
```

```python
accuracy_score(y_test , y_pred_svm)               # best accuracy score 
```

```python
confusion_matrix(y_test, y_pred_svm)
```

```python
print(classification_report(y_test, y_pred_svm))
```
